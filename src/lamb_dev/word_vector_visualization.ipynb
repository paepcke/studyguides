```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
vectors = {}
with open('/Users/andrewlamb/Google_Drive/Stanford/CS199/CompilersSelfPacedCS1/vectors.txt') as f:
    vocab_size, dimension = f.readline().strip().split()
    
    for line in f:
        tokens = line.strip().split()
        word = tokens[0]
        vector = [float(n) for n in tokens[1:]]
        vectors[word] = vector
        
print vocab_size, dimension
```

```python
X = np.array(vectors.values())
kmeans = KMeans(n_clusters=500)
clusters = kmeans.fit_predict(X)
```

```python
pca = PCA(n_components=2)
X_low_dim = pca.fit_transform(X)
```

```python
plt.scatter(X_low_dim[:,0], X_low_dim[:,1], c=clusters)
fig = plt.gcf()
ax = plt.gca()
for word, coord in zip(vectors.keys(), X_low_dim):
    ax.annotate(word, coord)
 
fig.set_size_inches(200, 200)
fig.savefig('scatter.png')
fig.set_size_inches(20, 20)
```
